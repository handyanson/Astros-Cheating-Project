# Player Analysis 
In this note book we look at the indvidual hitting stats for players who over saw over 500 pitches and had over 100 recorded bangs during the cheating period in 2017. To see more info about players and how frequently they cheated you can check out signstealingscandal.com

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import xticks
plt.rcParams['figure.figsize'] = (30,15)
```

```python
springer_df = pd.read_excel('spinger.xlsx')
bregman_df = pd.read_excel('bregman.xlsx')
beltran_df = pd.read_excel('beltran.xlsx')
gurriel_df = pd.read_excel('gurriel.xlsx')
correra_df = pd.read_excel('correra.xlsx')

```

```python
plt.rcParams['figure.figsize'] = (30,15)
plt.rc('legend', fontsize=20)
plt.plot(correra_df['Gtm'], correra_df['OBP'], '-', label = 'Carlos Correra')
plt.plot(springer_df['Gtm'], springer_df['OBP'], '-', label = 'George Springer')
plt.plot(beltran_df['Gtm'], beltran_df['OBP'], '-', label = 'Carlos Beltran')
plt.plot(gurriel_df['Gtm'], gurriel_df['OBP'], '-', label = 'Yuli Gurriel')
plt.plot(bregman_df['Gtm'], bregman_df['OBP'], '-', label = 'Alex Bregman')

plt.axvline(x=50, color = 'black',ls ='--')
plt.axvline(x=152, color = 'black',ls ='--')
xticks([3, 27,50, 56, 83, 107, 135, 152, 162], ['April', 'May', 'Cheating Begins', 'June', 'July', 'August', 'September','Cheating Ends', 'October'])
plt.title('Date vs. On Base % (OPB)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('OBP', fontsize = 20)
plt.legend()
```
